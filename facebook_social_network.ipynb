```python
from pyvis.network import Network
import pandas as pd
from itertools import cycle
import pyvis
```

```python
# Create a new Pyvis network
net = Network(notebook=True, cdn_resources="remote", width="100%", height="750px", bgcolor='#222222', font_color='white')

# Define node details
node_ids = [1, 2, 3, 4, 5]
labels = ["Fay", "Olivia", "Edison", "Jessi", "MAOMAO"]
colors = ['#057a36', '#0748b0', '#edc4f5', '#e64c05', '#940101']

# Add each node individually
for node_id, label, color in zip(node_ids, labels, colors):
    net.add_node(node_id, label=label, color=color)

# Add single edge
net.add_edge(1, 5,value=2)

# Add multiple edges
edges = [(2, 5), (2, 4), (1, 4), (2, 3), (1, 2)]
for source, target in edges:
    net.add_edge(source, target)
# Show the network
net.show("example_network.html")
```

```python
net = Network(
    notebook=True,
    cdn_resources="remote",
    bgcolor="#222222",
    font_color="white",
    height="750px",
    width="100%",
    #select_menu=True,
    #filter_menu=True,
)

nodes = set(sample['person1']).union(set(sample['person2']))
colors = ['#c1e1f5', '#f2cee0', '#7f47ad', '#71bd62', '#f7f760', '#b5f5d9','#094dad','#f5d5ea','#edc7ce']
color_cycle = cycle(colors)  # Use cycle to repeat colors if not enough
color_map = {node: colors[i % len(colors)] for i, node in enumerate(nodes)}

# Add nodes to the network with unique colors
for node in nodes:
    net.add_node(node, label=str(node), color=color_map[node])

# Add edges to the network
edges = sample.values.tolist()
for edge in edges:
    net.add_edge(edge[0], edge[1])


# Display the network
net.show_buttons(filter_="physics")
net.show("facebook_social.html")
```

```python
import json
```

```python
# Extract node data from the network
node_data = [
    {
        'id': node.get('id'),
        'label': node.get('label'),
        'color': node.get('color')
    }
    for node in net.nodes
]

# Extract edge data from the network
edge_data = [
    {
        'from': edge['from'],
        'to': edge['to'],
        'title': edge.get('title', ''),
        'value': edge.get('value', 1)  # Default to 1 if no specific value provided
    }
    for edge in net.edges
]

# Combine node and edge data into one dictionary for easy JSON serialization
network_data = {
    'nodes': node_data,
    'edges': edge_data
}

# Serialize and save to a JSON file
with open('network_data.json', 'w') as f:
    json.dump(network_data, f, indent=2)
```
